```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
north_america = pd.read_csv('https://raw.githubusercontent.com/deepaksood619/DS_ML/master/Pandas%20Concatenation%20Tutorial/north_america_2000_2010.csv', index_col=0)
south_america = pd.read_csv('https://raw.githubusercontent.com/deepaksood619/DS_ML/master/Pandas%20Concatenation%20Tutorial/south_america_2000_2010.csv', index_col=0)
```

```python
print(north_america)
print(south_america)
```

```python
north_america.plot()
```

```python
north_america.transpose().plot(title='Average Labor Hours Per Year')
print(north_america.transpose())
plt.show()
```

```python
south_america.transpose().plot(title='Average Labor Hours Per Year')
plt.show()
```

# Concatenating Americas data

```python
americas = pd.concat([north_america, south_america])
print(americas)
```

```python
americas_dfs = [americas]
for year in range(2011, 2016):
    filename = "https://raw.githubusercontent.com/deepaksood619/DS_ML/master/Pandas%20Concatenation%20Tutorial/americas_{}.csv".format(year)
    df = pd.read_csv(filename, index_col=0)
    americas_dfs.append(df)
print(americas_dfs[1])
```

```python
americas = pd.concat(americas_dfs, axis=1)
americas.index.names = ['Country']
print(americas)
```

```python
americas.transpose().plot(title='Average Labor Hours Per Year')
plt.show()
```

# Appending data from other continents

```python
con_list = ['asia','europe','south_pacific']
```

```python
con_dfs = []
for con in con_list:
    filename = "https://raw.githubusercontent.com/deepaksood619/DS_ML/master/Pandas%20Concatenation%20Tutorial/{}_2000_2015.csv".format(con)
    df = pd.read_csv(filename, index_col=0)
    con_dfs.append(df)
```

```python
con_df = pd.concat(con_dfs, axis=0)
americas.index.names = ['Country']
print((con_df))
```

```python
world = americas.append(con_df)
world.index
```

```python
world.transpose().plot(title='Average Labor Hours Per Year')
plt.show()
```

# Customizing the visualization

```python
world.transpose().plot(figsize=(10,10), colormap='gist_rainbow', linewidth=2, title='Average Labor Hours Per Year')
plt.legend(loc='right', bbox_to_anchor=(1.3, 0.5))
plt.show()
```

# Merging historical labor data

```python
historical = pd.read_csv('https://raw.githubusercontent.com/deepaksood619/DS_ML/master/Pandas%20Concatenation%20Tutorial/historical.csv', index_col=0)
historical.head()
```

```python
print("World rows & columns: ", world.shape)
print("Historical rows & columns: ", historical.shape)
```

```python
world_historical = pd.merge(historical, world, left_index=True, right_index=True, how='right')
print(world_historical.shape)
world_historical.head()
```

```python
world_historical = historical.join(world, how='right')
print(world_historical.head())
```

```python
world_historical.sort_index(inplace=True)
world_historical.transpose().plot(figsize=(15,10), colormap='rainbow', linewidth=2, title='Average Labor Hours Per Year')
plt.legend(loc='right', bbox_to_anchor=(1.15, 0.5))
plt.show()
```
